```python
import numpy as np
```

ARRAY OPERATIONS

```python
a = np.array([10,20,30])
b = np.array([1,2,3])

print(a+b)
print(a*b)
```

ZEROS AND ONES


```python
print(np.zeros((2,3)))
print(np.ones((3,3)))
```

MEAN, MAX, MIN

```python
data = np.array([10,20,30,40])

print("mean:", np.mean(data))
print("max:", np.max(data))
print("min:", np.min(data))
```
