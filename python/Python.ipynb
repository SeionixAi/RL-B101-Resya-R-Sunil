EVEN OR ODD

```python
num = int(input("Enter a number:"))
if num%2 ==0:
    print("even number")
else:
    print("odd number")
```

```python
numbers = [10, 20, 30, 40, 50]
print("sum:", sum(numbers))
print('Max:', max(numbers))
print("min:", min(numbers))
```
