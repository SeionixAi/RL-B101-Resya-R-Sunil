1. Dynamic Typing

* Assigns an integer value to a variable
* Reassigns a string value to the same variable
* Prints the value and its type at each step

```python
x = 10
print(x, type(x))
 
x = "Hello Python"
print(x, type(x))
```

2. Mutable vs Immutable Data Types

```python
my_list = [1,2,3]
my_list[0] = 10               # Lists are mutable
print(my_list)
```

```python
my_tuple = (1,2,3)
my_tuple[0] = 10         # Tuples are immutable
print(my_tuple)
```

3. Local and Global Variables

```python
x =100                 # global variable


def demo():
    y = 50             # local variable
    print(x,y)
demo()
```

4. Return Statement in a Function

```python
def add(a,b):
    return a + b

result = add(5,3)
print(result)
```

5. List Comprehension - Creates a list of squares from 1 to 10

```python
squares = [x*x for x in range(1,11)]
print(squares)
```

6. break, continue, and pass

```python
for i in range(1,11):
    if i == 5:
        continue
    if i ==5:
        break
    if i ==7:
        pass
    print(i)
```

7. Recursion - Write a recursive function that:
* Finds the factorial of a number

```python
def factorial(n):
    if n==0:
        return 1
    return n * factorial(n-1)
print(factorial(5))

```

8. Lambda Function - Write a lambda function that:
* Returns the square of a number

```python
square = lambda x: x * x
print(square(4))
```

9. is vs ==

```python
a = [1, 2, 3]
b = [1, 2, 3]
 
print(a == b)
print(a is b)
```
